- Calculadora Inteligente

```python
# Entrada do nome do usuário:
nome_usuario = input("Insira seu nome: ")
print(f"Olá {nome_usuario}! Bem-vindo a calculadora inteligente")

while True:
  #Escolhedo o tipo de operação:
  print("\nTipos de operações: ")
  print("1 - Soma")
  print("2 - Subtração")
  print("3 - Multiplicação")
  print("4 - Divisão")

  operacao = input(f"\n{nome_usuario}, qual tipo de operação gostaria de realizar? ")

  #Entrada dos números para a realização da operação:
  Num_1 = float(input("Digite o primeiro número: "))
  Num_2 = float(input("Digite o segundo número: "))

  #Condicionais para cada tipo de operação:
  if operacao == "1":
    resultado = Num_1 + Num_2
    print("O resultado da soma é ", resultado)

  elif operacao == "2":
    resultado = Num_1 - Num_2
    print("O resultado da subtração é ", resultado)

  elif operacao == "3":
    resultado = Num_1 * Num_2
    print("O resultado da multiplicação é ", resultado)

  elif operacao == "4":
      if Num_2 != 0:
        resultado = Num_1 / Num_2
        print("O resultado da divisão é ", resultado)
      else:
        print("Não é possível dividir por zero")

  else:
    print("Operação inválida.")

  # Repetição da calculadora:
  repetir = input (f"\nDeseja realizar outra operação, {nome_usuario}? \nDigite s para Sim ou n para Não:")
  if repetir.lower() != "s":
    print("Fim da operação.")
    break
```
